```python
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
%matplotlib inline
```

```python
adc_path = Path('./nifti/ProstateX-0000/adc/7_ep2d_diff_tra_dyndist_adc.nii.gz')
adc = sitk.ReadImage(str(adc_path))
```

```python
bval_path = Path('./nifti/ProstateX-0000/bval/8_ep2d_diff_tra_dyndistcalc_bval.nii.gz')
bval = sitk.ReadImage(str(adc_path))
```

```python
ktrans_path = Path('./nifti/ProstateX-0000/ktrans/ProstateX-0000-ktrans.nii.gz')
ktrans = sitk.ReadImage(str(ktrans_path))
```

```python
t2_path = Path('./nifti/ProstateX-0000/t2/4_t2_tse_tra.nii.gz')
t2 = sitk.ReadImage(str(t2_path))
```

```python
sitk_images = {'adc':adc, 'bval':bval, 'ktrans':ktrans, 't2':t2}
```

```python
# https://stackoverflow.com/questions/47900453/how-to-get-a-histogram-of-3d-images
# https://stackoverflow.com/questions/34482154/simpleitk-how-to-get-maximum-and-minimum-intensity-in-image

def generateHistograms(sitk_images):
    for key, value in sitk_images.items():
        sitk_image = value
        title = key
        
        image_np_array = sitk.GetArrayViewFromImage(sitk_image)
        flat_image = image_np_array.flatten()
        n, bins, patches = plt.hist(flat_image, bins=256, range= (1,flat_image.max()), color ='gray', histtype = 'stepfilled')
        plt.title(title + ' histogram')
        plt.show()
        plt.cla()
        plt.clf()
```

```python
generateHistograms(sitk_images)
```
